# Matplotlib

```python
import matplotlib
```

```python
from sklearn.datasets import make_regression
# Gerando amostras iniciais com o metodo de regressao do sklearn
# 200 exemplos - caracteristicas envolvendo 1 unidade, taxa de erro 30
x, y = make_regression(n_samples=200, n_features=1, noise=30)
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.show()
```
